two strategies:
1) pick a door and stick with it
2) pick a door, let MH show you a goat, and then pick the remaining

```julia
MH_data = rand(1:3,10)
s1 = []
s2 = []
for run in MH_data
    print(run)
    s1_pick = rand(1:3)
    if s1_pick == run
        push!(s1,1)
    else
        push!(s1,0)
    end
    s2_pick = rand(1:3)
    if s2_pick == run
        push!(s2,0)
    else
        push!(s2,1)
    end
end
mean(s1),mean(s2)
```
